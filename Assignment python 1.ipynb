#Compare and contrast mutable and immutable in python with example?
- In Python, mutable and immutable are two types of objects that can be used to
  store and manipulate data.
  (i) Definition: Mutable objects are those that can be modified after creation.
  Examples: Lists, dictionaries, sets, and user-defined objects (ii) Definition: Immutable objects are those that cannot be modified after creation.Examples: Integers, floats, strings, tuples, and frozensets.
                           Comparision
   | Modifiability | Can be modified | Cannot be modified |
| Examples | Lists, dictionaries, sets | Integers, floats, strings, tuples |
| Identity | Unique identity | Unique identity |
| Methods | Various methods to modify | No methods to modify |

```python
#mutable object
my_list=[1,2,3]
my_list.append(4)
print(my_list)
```

```python
#immutable object
my_string="Hello"
new_string=my_string+"World"
print(new_string)
```

#Explain the concept of type casting in Python with examples?
- In Python, type casting is the process of converting an object of one data type to another data type. This can be done explicitly or implicitly. 1 Implicit Type Casting:-
Implicit type casting occurs automatically when Python interpreter encounters an expression with different data types. Python automatically converts the data type of one operand to match the data type of the other operand.2 Explicit Type Casting:-
Explicit type casting is done manually using built-in functions like int(), float(), str(), etc.


```python
#implicit type Casting
a=4
b=5.5
print(a+b)
```

```python
#Explicit type casting
a=5.5
b=int(a)
print(b)
```

#How do conditional statements work in Python? Illustrate with examples?
- Conditional statements in Python are used to execute different blocks of code based on certain conditions. They allow your program to make decisions and take different paths depending on the input or situation.1:-If Statement
The if statement is used to execute a block of code if a certain condition is true.

```python
a=5
if a<10:
  print('a is less than 10')
```

(ii)If-Else Statement
The if-else statement is used to execute one block of code if a certain condition is true and another block of code if the condition is false.

```python
a=5
if a<10:
  print('a is less than 10')
else:
  print('a is greater than 10')
```

(iii)If-Elif-Else Statement
The if-elif-else statement is used to check multiple conditions and execute different blocks of code based on those conditions.

```python
a=5
if a<10:
  print('a is less than 10')
elif a==10:
  print('a is equal to 10')
else:
  print('a is greater than 10')
```

#Describe the different types of loops in Python and their use cases with examples?
- Python has two primary types of loops: for loops and while loops. (i)
For Loops:-
for loops are used to iterate over a sequence (such as a list, tuple, or string) and execute a block of code for each item in the sequence.

```python
#for loops
fruits=['Banana','Apple','Orange']
for Fruit in fruits:
    print(Fruit)
```

(ii)While Loops:-

while loops are used to execute a block of code as long as a certain condition is true.

```python
i=0
while i<5:
  print(i)
  i+=1
```
